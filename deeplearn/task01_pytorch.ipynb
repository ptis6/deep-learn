```python
import torch
import torch.nn as nn
import numpy as np
print(1)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
print(1)
```

```python
input_size = 1
output_size = 1
num_epochs = 60
learning_rate = 0.001

# Toy dataset
x_train = np.array([[1,2],[3,4],[5,6]], dtype=np.float32)

y_train = np.array([[1],[2],[3]], dtype=np.float32)

# Linear regression model
model = nn.Linear(input_size, output_size)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)  

# Train the model
for epoch in range(num_epochs):
    # Convert numpy arrays to torch tensors
    inputs = torch.from_numpy(x_train)
    targets = torch.from_numpy(y_train)

    # Forward pass
    outputs = model(inputs)
    loss = criterion(outputs, targets)
    
    # Backward and optimize
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 5 == 0:
        print ('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))

# Plot the graph
```

```python
class Linear_class(nn.Module):
    def __init__(self):
        super(Linear_class, self).__init__()
        self.linear=nn.Linear(2,1)
    def forward(self,x):
        out= self.linear(x)
        return out

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
model=Linear_class()

model.to(device)
```

```python

x_train = np.array([[1,2],[3,4],[5,6]], dtype=np.float32)

y_train = np.array([[1],[2],[3]], dtype=np.float32)
epochs=100000
learing_rate=0.01

opt=torch.optim.SGD(model.parameters(),lr=learing_rate)

criterion=nn.MSELoss()


for epoch in range(epochs):
    epoch+=1
    inputs=torch.from_numpy(x_train)
    labels=torch.from_numpy(y_train)
    inputs,labels=inputs.to(device),labels.to(device)
    
    y_predict=model(inputs)


    opt.zero_grad()

    loss=criterion(y_predict,labels)
    loss.backward()

    opt.step()

    if epoch %1000==0:
        print("epoch：{},loss：{}".format(epoch,loss))
```

```python
predicted = model(torch.from_numpy(x_train)).detach().numpy()
plt.plot(x_train, y_train, 'ro', label='Original data')
plt.plot(x_train, predicted, label='Fitted line')
plt.legend()
plt.show()
```
